```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x,y, 'r*')
plt.plot(x, y)
plt.title("Squared relation")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
```

**OOP Approach**

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1, 0.8,0.8])
axes2 = fig.add_axes([0.23, 0.45,0.35,0.35])
axes.set_xlabel("Values of X")
axes.set_ylabel("Values of Y")
axes.set_title('Squared Relations')
axes.plot(x,y)
axes2.set_xlabel("Values of X")
axes2.set_ylabel("Values of Y")
axes2.set_title('Squared Relations')
axes2.plot(x,y, 'g-')
```
